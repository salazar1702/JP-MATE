```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de x
x = np.linspace(-10,10,200)

# Función
y = x**2

# Derivada
dy = 2*x

# Gráficas
plt.plot(x,y,label="f(x)=x²")
plt.plot(x,dy,label="f'(x)=2x")

plt.title("Función y su derivada")
plt.xlabel("x")
plt.ylabel("y")

plt.grid()
plt.legend()

plt.show()
```
